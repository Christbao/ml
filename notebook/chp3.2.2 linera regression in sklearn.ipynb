## 稳健线性回归

我们分析世界，需要对世界建模，把世界中的现象抽象成模型。每个模型，又存在一些参数，通过调节参数，可以得到不同的实例，进行推演。我们观察现象，得到一堆数据。如何为这堆数据找一个合适的模型，再确定合适的模型参数，这是很重要的问题，是人类理性的基础。

数据分两种：有效数据（inliers）和无效数据（outliers）。那些偏差不大的数据是有效数据，偏差大的数据是无效数据。
如果有效数据占大多数，无效数据只是很少量时，我们可以通过最小二乘法或类似的方法来确定模型的参数和误差。如果无效数据很多（比如，超过了50%的数据是无效数据），最小二乘法就失效了，我们需要新的算法。

线性模型很容易受outlier影响，因此提出了稳健线性回归。比如RANdom SAmple Consensus (RANSAC-随机抽样一致)。RANSAC算法的输入是一组观测数据，一个可以解释或者适应于观测数据的参数化模型，一些可信的参数。RANSAC通过反复选择数据中的一组随机子集来达成目标。被选取的子集被假设为局内点，并用下述方法进行验证：

1. 有一个模型适应于假设的inlier点，即所有的未知参数都能从假设的inlier点计算得出。
2. 用1中得到的模型去测试所有的其它数据，如果某个点适用于估计的模型，认为它也是inlier点。
3. 如果有足够多的点被归类为假设的inlier点，那么估计的模型就足够合理。
4. 然后，用所有假设的inlier点去重新估计模型
5. 最后，通过估计inlier点与模型的错误率来评估模型。

这个过程被重复执行固定的次数，每次产生的模型要么因为inlier点太少而被舍弃，要么因为比现有的模型更好而被选用

http://en.wikipedia.org/wiki/RANSAC

```python
from sklearn.linear_model import RANSACRegressor
ransac = RANSACRegressor(LinearRegression(),max_trials=100,min_samples=50,
                         residual_metric=lambda x: np.sum(np.abs(x), axis=1),residual_threshold=5.0,random_state=0)
ransac.fit(X, y)
```

```python
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
line_X = np.arange(3, 10, 1)
line_y_ransac = ransac.predict(line_X[:, np.newaxis])
plt.scatter(X[inlier_mask], y[inlier_mask],c='blue', marker='o', label='Inliers')
plt.scatter(X[outlier_mask], y[outlier_mask],c='
            ', marker='s', label='Outliers')
plt.plot(line_X, line_y_ransac, color='red')
plt.xlabel('Average number of rooms [RM]')
plt.ylabel('Price in $1000\'s [MEDV]')
plt.legend(loc='upper left')
```

```python
print(ransac.estimator_.score(X,y))
print(ransac.estimator_.coef_[0],ransac.estimator_.intercept_)
print("price for 5 rooms:{}".format(ransac.estimator_.coef_[0]*5+ransac.estimator_.intercept_))
```

## 评估模型

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
```

```python
X_train.shape,X_test.shape
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test)
```

```python
plt.scatter(y_train_pred, y_train_pred - y_train,c='blue', marker='o', label='Training data')
plt.scatter(y_test_pred, y_test_pred - y_test,c='red', marker='s', label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-10, xmax=50, lw=2, color='red')
plt.xlim([-10, 50])
```

## polynomial regression

$$ y= w_0+w_1x+w_2x^2+...+w_dx^d$$

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, PolynomialFeatures
%matplotlib inline
```

```python
X = np.array([258.0, 270.0, 294.0,320.0, 342.0, 368.0,396.0, 446.0, 480.0,586.0,550])[:, np.newaxis]
y = np.array([236.4, 234.4, 252.8,298.6, 314.2, 342.2,360.8, 368.0, 391.2,390.8,420])
```

```python
X.shape
```

```python
lr = LinearRegression()
pr = LinearRegression()
quadratic = PolynomialFeatures(degree=2)
X_quad = quadratic.fit_transform(X)
```

```python
lr.fit(X, y)
X_fit = np.arange(250,600,10)[:, np.newaxis]
y_lin_fit = lr.predict(X_fit)
pr.fit(X_quad, y)
y_quad_fit = pr.predict(quadratic.fit_transform(X_fit))
plt.scatter(X, y, label='training points')
plt.plot(X_fit, y_lin_fit,label='linear fit', linestyle='--')
plt.plot(X_fit, y_quad_fit,label='quadratic fit')
plt.legend(loc='upper left')
```

```python
print("training score:{:.3f}".format(lr.score(X,y)))
print("training score:{:.3f}".format(pr.score(X_quad,y)))
```

```python
lr.fit(X[:8], y[:8])
X_fit = np.arange(250,600,10)[:, np.newaxis]
y_lin_fit = lr.predict(X_fit)
pr.fit(X_quad[:8], y[:8])
y_quad_fit = pr.predict(quadratic.fit_transform(X_fit))
plt.scatter(X[:8], y[:8], label='training points')
plt.scatter(X[8:], y[8:], label='testing points',c="r")
plt.plot(X_fit, y_lin_fit,label='linear fit', linestyle='--')
plt.plot(X_fit, y_quad_fit,label='quadratic fit')
plt.legend(loc='upper left')
```

```python
print("training score:{:.3f}".format(lr.score(X,y)))
print("training score:{:.3f}".format(pr.score(X_quad,y)))
```

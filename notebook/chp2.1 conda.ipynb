## 使用conda

+ 获取帮助:

`conda help`  
`conda update --help`

+ 检查conda版本  

`conda --version`

+ 升级conda

`conda update conda`

+ 列出已经安装的包

`conda list`

+ 安装包

`conda install packagename`  
`conda install numpy pandas scikit-learn matplotlib seaborn`

[conda命令](http://conda.pydata.org/docs/index.html)  
http://conda.pydata.org/docs/index.html

+ conda 镜像

`conda config --add channels https://mirrors.tuna.tsinghua.edu.cn/anaconda/pkgs/free/`  
`conda config --set show_channel_urls yes`

## 管理conda environments

	conda env --help

### create

	conda create --name yournames packages-your-want
	conda create -n yournames packages-your-want

### change env

	source activate yourname
	activate your name
	source deactivate 
	deactivate your

### set python version

	conda create -n yourname python=x.x packages-your-want


### 当前environments 信息 list env

	conda info --envs

### clone env

	conda create -n name1 --clone name2

### remove env

	conda remove --name yourname --all
	conda info --envs

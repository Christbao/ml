```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
df=pd.read_csv("../data/seperator_class.csv")
X,y=df[[0,1]].values,df[[2]].values.ravel()
```

```python
logi=LogisticRegression()
model=logi.fit(X,y)
```

```python
cancer=load_breast_cancer()
X_train,X_test,y_train,y_test=train_test_split(cancer.data,cancer.target,stratify=cancer.target,random_state=0)
logreg=LogisticRegression().fit(X_train,y_train) #default C=1
print("training set score:{:.3f}".format(logreg.score(X_train,y_train)))
print("test set score:{:.3f}".format(logreg.score(X_test,y_test)))
```

```python
logreg100=LogisticRegression(C=100).fit(X_train,y_train) #default C=1,C越大模型越灵活
print("training set score:{:.3f}".format(logreg100.score(X_train,y_train)))
print("test set score:{:.3f}".format(logreg100.score(X_test,y_test)))
```

```python
logreg001=LogisticRegression(C=0.01).fit(X_train,y_train) #default C=1,C越大模型越灵活,同时缺省是L2 regularization
print("training set score:{:.3f}".format(logreg001.score(X_train,y_train)))
print("test set score:{:.3f}".format(logreg001.score(X_test,y_test)))
```

```python
plt.figure(figsize=(12,9))
plt.plot(logreg.coef_.T,'s',label="C=1")
plt.plot(logreg100.coef_.T,'^',label="C=100")
plt.plot(logreg001.coef_.T,'v',label="C=0.01")

plt.xticks(range(cancer.data.shape[1]),cancer.feature_names,rotation=90)
plt.xlabel("coefficient index")
plt.ylabel("coefficient results")
plt.hlines(0,0,cancer.data.shape[1])
plt.ylim(-5,5)
plt.legend();
```

```python
plt.figure(figsize=(12,9))
for C,marker in zip([0.001,1,100],['o','^','v']):
    lr_l1=LogisticRegression(C=C,penalty="l1").fit(X_train,y_train)
    print("training accuracy with l1,C={:.3f}:{:.3f}".format(C,lr_l1.score(X_train,y_train)))
    print("testing accuracy with l1,C={:.3f}:{:.3f}".format(C,lr_l1.score(X_test,y_test)))
    plt.plot(lr_l1.coef_.T,marker,label="C={:.3f}".format(C))

plt.xticks(range(cancer.data.shape[1]),cancer.feature_names,rotation=90)
plt.xlabel("coefficient index")
plt.ylabel("coefficient results")
plt.hlines(0,0,cancer.data.shape[1])
plt.ylim(-5,5)
plt.legend(loc=1);
```
